```python
df['Relative Strength'] = df['Estimated 1RM'] / df['bodyweight']
```

```python
df["Gym Type"]=np.where(df['date']>='2025-12-13','Soro Gym','College Gym')
```

```python
df[df['date']>='2025-12-13']
```

```python
df['Week']=df['date'].dt.isocalendar().week
```

```python
df['Month']=df['date'].dt.strftime('%b')
```

```python
df.columns=df.columns.str.upper()
```

```python
df['EXERCISE']=df['EXERCISE'].str.title()
df['MUSCLE GROUP']=df['MUSCLE GROUP'].str.title()
```

```python
df["PHASE"] = np.where(
    (df["DATE"] >= "2025-04-06") & (df["DATE"] <= "2025-06-14"),
    "Bulk",
    np.where(
        (df["DATE"] >= "2025-06-15") & (df["DATE"] <= "2025-11-11"),
        "Cut",
        np.where(
            df["DATE"] >= "2025-11-12",
            "Bulk",
            "Unknown"
        )
    )
)
```

```python
df=df[['DATE','WEEK','MONTH','GYM TYPE','BODYWEIGHT','PHASE','EXERCISE','MUSCLE GROUP','WEIGHT','REPS','VOLUME','ESTIMATED 1RM','RELATIVE STRENGTH']]
```
